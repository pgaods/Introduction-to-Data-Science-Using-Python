In this lecture we study how to scrape Yelp reviews:

```python
import webbrowser
import requests

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import bs4
import re 

# pd.options.display.max_rows = 999
```

```python
res = requests.get('https://www.cms.gov/icd10m/version37-fullcode-cms/fullcode_cms/P1559.html')
#res = requests.get('https://www.cms.gov/icd10m/version37-fullcode-cms/fullcode_cms/P1931.html')

try:
    res.raise_for_status()
except Exception as exc:
    print('There was a problem: %s' % (exc))
soup = bs4.BeautifulSoup(res.text, 'html.parser')
```

```python
data = soup.find_all("table")[2]
meat = data.find_all('tr')
finallist=[td.text.strip() for row in meat for td in row.find_all('td')]

finallist2 = np.array(finallist).reshape(-1, 4)
df=pd.DataFrame(finallist2, columns=['proccode', 'MDC', 'MSDRG', 'Surgical Category'])
df = df.apply(lambda x: x.str.strip()).replace('', np.nan) # replacing blank space with np.nan
df.head(9)
```

```python
df['proccode']=df['proccode'].fillna(method="ffill") # to carry the values forward
df.head(9)
```
