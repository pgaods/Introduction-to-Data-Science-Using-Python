In this lecture, we give a brief introduction to Anaconda's virtual environment and other relevant information regarding Anaconda distribution. Simply put, virtual environments in Anaconda allow users to set up virtual installations of Python and associated libraries on the computer. One can have multiple versions of Python or libraries and easily activate or deactivate these environments. 

Here is an example: suppose we have a Python library used in production. A new version of the library was released and we don't want or old code to break while we still want to explore the updated library. In this case we need virtual environment. Another example is that sometimes we need both Python 2.7 version and Python 3.5.

There is the 'virtualenv' library for normal Python distributions that can handle multiple environment. However, Anaconda has a built-in virtual environment manager that can make this very smooth and easy. A useful resource link can be found here: http://conda.pydata.org/docs/using/envs.html. The idea is that by using 'Command Prompt' on your computer, you can choose to create/activate/deactivate/invoke different Python environments. For example, many companies use this to achieve version control of their production codes.

Pipenv is often used in conjunction with git. We will not elaborate on this for now. The third way is to use Anaconda's navigator. ### I. Fundamentas of Creating Virtual Environments Using Anaconda Prompt

A **conda environment** is a directory that contains a specific collection of conda packages that you have installed. For example, you may have one environment with Numpy 1.7 and its dependencies, and another environment with Numpy 1.6 for legacy testing. If you change one environment, your other environments are not affected. You can easily activate or deactivate environments, which is how you switch between them. You can also share your environment with someone by giving them a copy of your environment "_.yaml_" file. We will cover this topic later. 

In the anaconda prompt, to create an environment, you can type in the following command (you need to replace _myenv_ with a user-defined environment name):

   - conda create --name _myenv_
   
Now let's create an environment called 'myenv':

```python
from IPython.display import Image
Image("creating a conda env.PNG")
```

As the instruction said, to activate or deactivate the environment, here are the commands:

   - conda activate _myenv_
   - conda deactivate
   
To create an environment with a specific package, here is what you can do:

   - conda create -n _myenv_ _packagename_
   - conda create -n _myenv_ _packagename=version-number_

Here is an example:

   - conda create -n myenv scipy=0.15.0
   
Here is an example of creating an environment with a specific version of Python and multiple packages:

   - conda create -n myenv python=3.6 scipy=0.15.0 astroid babel
   
An important tip is that we should always try to install all the programs that we want in this environment at the same time. Installing one program at a time can lead to dependency conflicts.

You can control where a conda environment lives by providing a path to a target directory when creating the environment. For example, suppose you already have had an environment called 'myenvs'. And we want to create a sub-environment in this folder. The following command will create a new environment in a subdirectory called 'subenvs'. We first get to the myenv folder and then create the 'subenvs' environment:

   - cd .\anaconda3\envs\myenv
   - conda create --prefix ./subenvs numpy=1.16
   
You then activate an environment created with a prefix using the same command used to activate environments created by name:

   - conda activate C:\Users\pgao\anaconda3\envs\myenv\subenvs
   
Specifying a path to a subdirectory of your project directory when creating an environment has the following benefits:

   1. It makes it easy to tell if your project uses an isolated environment by including the environment as a subdirectory.
   2. It makes your project more self-contained as everything, including the required software, is contained in a single project directory.

In general, creating sub-environment might be tricky in terms of folder structure. We don't recommend using it unless it's totally necessary because sometimes it might be easy to mess up the folder structure and it's hard to remove it. 

To remove an environment called 'myenv', here what we can do:

   - conda remove --name myenv --all
   
To know how many envinroments are there altogether, we can do:

   - conda info --env

To see a list of all packages installed in a specific environment:

If the environment is not activated, in your terminal window or an Anaconda Prompt, run:

  - conda list -n _myenv

If the environment is activated, in your terminal window or an Anaconda Prompt, run:

  - conda list

To see if a specific package is installed in an environment, in your terminal window or an Anaconda Prompt, run:

  - conda list -n _myenv_ _package-name_

### III. Conda Packages

A conda package is a compressed tarball file (.tar.bz2) or conda (.conda file) that contains:

   - system-level libraries
   - Python or other modules
   - executable programs and other components
   - metadata under the 'info/' directory
   - a collection of files that are installed directly into an _install_ prefix

Conda keeps track of the dependencies between packages and platforms. The conda package format is identical across platforms and operating systems. Only files, including symbolic links, are part of a conda package. Directories are not included. Directories are created and removed as needed, but you cannot create an empty directory from the tar archive directly.

In general, package structures look like the following:

```python
from IPython.display import Image
Image("conda package structure.PNG")
```

You may search for packages using the following command:

   - conda search _package-name_
   
Installing the package is easy as well, just change the command above from 'search' to install. Make sure when you install you are in the right environment. 

Updating packages in Python is easy too. There are certain types of packages that are constantly being updated (say the machine learning library). In this case, one can go to the command prompt and type in the following command:

   - conda update _your_package_name_
   
**Conda channels** are the locations where packages are stored. They serve as the base for hosting and managing packages. Conda packages are downloaded from remote channels, which are URLs to directories containing conda packages. The conda command searches a default set of channels, and packages are automatically downloaded and updated from https://repo.anaconda.com/pkgs/. You can modify what remote channels are automatically searched. We will not cover the details here. 

### IV. Sharing Environments Using the "yml" File

You may want to share your environment with someone else---for example, so they can re-create a test that you have done. To allow them to quickly reproduce your environment, with all of its packages and versions, give them a copy of your '_yourenvironment_.yml' file. To share environment, you need to go through 2 steps. First, activate your current environment. Second, export your active environment to a new file using the code below:

   - conda env export > _your-environment_.yml
